```python

import pandas as pd
import numpy as np 

```

```python

def update():
    a = pd.DataFrame()
    for i in ["May", "June", "July", "August", "September", "October"]:
        df = pd.read_csv(f'data/matrices/{i}_Data_Matrix (1).csv', encoding ="latin1")
        df["month"] = i
        a = pd.concat([a,df])
    
    a.to_csv("data/Data_Matrix (1).csv", index = False)
update()
```

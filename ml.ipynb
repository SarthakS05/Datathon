```python

import pandas as pd
import numpy as np 

```

```python

def update_monthly():
    a = pd.DataFrame()
    for i in ["May", "June", "July", "August", "September", "October"]:
        df = pd.read_csv(f'data/matrices/{i}_Data_Matrix (1).csv', encoding ="latin1")
        df["month"] = i
        a = pd.concat([a,df])
    a["Amount"] = (
        a["Amount"]
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )
    a["Count"] = a["Count"].str.replace(",", "", regex = False)
    a["priceper"] = (a["Amount"].astype(float)/a["Count"].astype(float)).round(2)
    print(a)
    a.to_csv("data/Data_Matrix.csv", index = False)
update_monthly()
```

```python
def ingredients():
    ing = pd.read_csv("data/MSY Data - Ingredient.csv")
    dat = pd.read_csv("data/Data_Matrix.csv")

    extr = dat.merge(ing, left_on=  "Item Name", right_on = "Item name", how = "left")

    extr.to_csv("data/extr.csv", index = False)
    return extr
df = ingredients()
print(ingredients().columns)
```
